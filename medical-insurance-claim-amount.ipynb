**Dataset:** [Medical Cost Personal Dataset](https://www.kaggle.com/datasets/mirichoi0218/insurance)

```python
# Encode categorical variables: sex and smoker (binary), region (one-hot)
df["sex"] = df["sex"].map({"female": 0, "male": 1})
df["smoker"] = df["smoker"].map({"no": 0, "yes": 1})
df = pd.get_dummies(df, columns=["region"], drop_first=True, dtype=int)

print("✅ Categorical variables encoded.")
print(f"Columns after encoding: {list(df.columns)}")
df.head()
```

## 📝 Conclusion & Key Insights

- **Age, BMI, and Smoking Status** are the most influential factors affecting insurance charges. Smokers and individuals with higher BMI or older age tend to have significantly higher medical costs.
- **Number of Children and Region** have a smaller but noticeable impact on charges, as seen in the boxplots.
- The **Linear Regression model** provides a reasonable estimate of insurance charges, with MAE and RMSE indicating the average prediction error in dollars.
- **Residual analysis** shows that while most predictions are close to actual values, there are some outliers, especially for very high charges.
- **Feature coefficients** confirm that being a smoker and having a higher BMI or age increases predicted charges the most.

### Recommendations:
- Insurance companies should pay special attention to applicants who are smokers, older, or have a high BMI, as they represent higher risk.
- Further improvements could be made by exploring non-linear models or adding more features.
